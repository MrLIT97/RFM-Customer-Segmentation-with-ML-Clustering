```python
#putting date in the standard format
df['transaction_date'] = pd.to_datetime(df['transaction_date'])
```

```python
#extracting needed subset
seg = df[['customer_id','transaction_id','transaction_date','list_price','standard_cost']].sort_values('customer_id',ascending=True)
seg
```

```python
#calculating recency by customer
today = pd.Timestamp.today()
today = pd.to_datetime('2024-12-06')

recency = seg.groupby('customer_id').agg({'transaction_date': lambda x:(today-x.max()).days})
```

```python
recency
```

```python
#calculating frequency by customer
freq = seg.drop_duplicates(subset=['customer_id','transaction_date']).groupby('customer_id')\
.agg({'transaction_date': lambda x:x.count()})
freq
```

```python
seg[['customer_id','transaction_date']].groupby('customer_id').count()
```

```python
#calculating profit
seg['profit'] = seg['list_price']-seg['standard_cost']
```

```python
seg
```

```python
#calculating monetary=profit by customer
monetary = seg[['customer_id','profit']].groupby('customer_id').mean()
monetary
```

```python
#creating RFM table
RFM = pd.concat([recency,freq,monetary], axis=1)
RFM.columns = ['recency', 'frequency','monetary']
```

```python
RFM
```

## Clustering

```python
#Scaling
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaled_rfm = scaler.fit_transform(RFM)
```

```python
scaled_rfm
```

```python
#Identifying number of clusters with elbow method
import warnings
warnings.filterwarnings('ignore')
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

coef = []
for i in range(1,11):
    kmeans = KMeans(n_clusters=i, random_state=42)
    kmeans.fit(scaled_rfm)
    coef.append(kmeans.inertia_)
plt.plot(range(1,11),coef,marker='o')
plt.show()
```

**Best number of clusters is 4, equivalent to the elbow joint**

```python
#training the model
kmeans = KMeans(n_clusters=4, random_state=42)
kmeans.fit(scaled_rfm)
kmeans.labels_
```

```python
#adding clusters
RFM['clusters'] = kmeans.labels_
RFM
```

```python
#grouping and sorting by profitability
RFM.groupby('clusters').mean().sort_values('monetary')
```

```python
#introducing the spending class segmentation
def spending_class(data):
    if data['clusters'] == 2:
        return 'LowSpender'
    elif data['clusters'] ==0:
        return 'MidSpender'
    elif data['clusters'] == 1:
        return 'AvgSpender'
    else:
        return 'HighSpender'
RFM['spending class'] = RFM.apply(spending_class, axis = 1)
```

```python
RFM
```

```python
#dataset with corresponding segmentation
customer = df.iloc[:,:17].merge(RFM[['recency','frequency','monetary','spending class']], on='customer_id', how='right')
customer
```

```python
#fixing the DOB
customer['DOB'] = pd.to_datetime('1900-01-01') + pd.to_timedelta(customer['DOB'], unit='D')
customer
```

```python
#Turing DOB to Age
customer['DOB'] = customer.agg({'DOB': lambda x: (today.year - x.year)})
customer.rename(columns={"DOB":"Age"})
```

```python
#extracting unique customer dataset with segmentation
customer = customer.drop_duplicates('customer_id', ignore_index=True)
customer
```
